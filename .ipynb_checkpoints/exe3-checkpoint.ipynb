```python
def calculate_vif_(df, thresh=5):
    '''
    Calculates VIF each feature in a pandas dataframe
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param thresh: the max VIF value before the feature is removed from the dataframe
    :return: dataframe with features removed
    '''
    const=add_constant(df)
    cols=const.columns
    variables=np.arange(const.shape[1])
    vif_df=pd.Series([variance_inflation_factor(const.values, i) 
               for i in range(const.shape[1])], 
              index=const.columns).to_frame()

    vif_df=vif_df.sort_values(by=0, ascending=False).rename(columns={0: 'VIF'})
    vif_df=vif_df.drop('const')
    vif_df=vif_df[vif_df['VIF'] > thresh]

    print('Features above VIF threshold:\n')
    print(vif_df[vif_df['VIF'] > thresh])

    col_to_drop = list(vif_df.index)

    return col_to_drop
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
pd.set_option('display.max_columns', 300)


from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

from sklearn import preprocessing
seed=42
```

```python
movies_df=pd.read_excel("raw/dataset_netflix.xlsx")
#diretores indicados ao oscar
director_list = [x.strip().lower() for x in open("directors.txt", "r") ]

print(movies_df.shape)
movies_df.sample(2)
```

```python
#genero
movies_df['listed_in']=movies_df['listed_in'].str.lower()
movies_df[['genre1','genre2','genre3']] =movies_df['listed_in'].str.split(',', expand=True)

#diretor
movies_df['director']=movies_df['director'].str.lower()
movies_df[['director1','director2','director3']] = movies_df['director'].fillna('unknow').str.split(',', expand=True).iloc[:,0:3]

#país
movies_df['country']=movies_df['country'].str.lower()
movies_df[['country1','country2']] = movies_df['country'].fillna('unknow').str.split(',', expand=True).iloc[:,0:2]

#data
movies_df.date_added=movies_df.date_added.fillna('1777-01-01 00:00:00')
movies_df.date_added=pd.to_datetime(movies_df.date_added, format='%Y-%m-%d %H:%M:%S')
movies_df['years_to_add']=movies_df.date_added.dt.year-movies_df.release_year
movies_df['years_to_add']=movies_df['years_to_add'].apply(lambda x: -1 if x<0 else x)

#diretores indicados ao oscar, fonte wikipedia
movies_df['oscar']=movies_df['director1'].apply(lambda x: 1 if any(i in x for i in director_list) else 0)
movies_df['oscar2']=movies_df['director2'].fillna('None').apply(lambda x: 1 if any(i in x for i in director_list) else 0)
movies_df['oscar_number']=movies_df['oscar']+movies_df['oscar2']

#duration_seasons
movies_df['duration_min'] = movies_df['duration'].apply(lambda x: str(x).replace(' min', '') if 'min' in str(x) else 0 ).astype('int64')
movies_df['duration_season']= movies_df['duration'].apply(lambda x: str(x).replace(' Seasons', '').replace(' Season', '') if 'Season' in str(x) else 0 ).astype('int64')
```

```python
"""
preprocessing
"""
movies_df.drop(movies_df.index[2017], inplace=True)
movies_df=movies_df[movies_df['rating']>0]
movies_clean_df=movies_df[['rating','type','release_year','genre1','years_to_add','oscar_number','duration_min','duration_season']]
dummies_df=pd.get_dummies(data=movies_clean_df, columns=['type', 'genre1'])
```

## Método paramétrico

```python
calculate_vif_(movies_df[['release_year','oscar_number','duration_min','duration_season']])

"""
VIF não mostrou nenhum colineariedade
"""
```

```python
from sklearn.model_selection import train_test_split

columns = [ 'years_to_add', 'oscar_number',
'duration_min', 'duration_season', 'type_Movie',
'genre1_action & adventure', 'genre1_anime features',
'genre1_anime series', 'genre1_british tv shows',
'genre1_children & family movies', 'genre1_classic & cult tv',
'genre1_classic movies', 'genre1_comedies', 'genre1_crime tv shows',
'genre1_cult movies', 'genre1_documentaries', 'genre1_docuseries',
'genre1_dramas', 'genre1_horror movies', 'genre1_independent movies',
'genre1_international movies', 'genre1_international tv shows',
'genre1_movies', 'genre1_music & musicals',
'genre1_reality tv', 'genre1_romantic movies',
'genre1_romantic tv shows', 'genre1_sci-fi & fantasy',
'genre1_spanish-language tv shows', 
'genre1_stand-up comedy', 'genre1_stand-up comedy & talk shows',
'genre1_thrillers', 'genre1_tv action & adventure',
'genre1_tv comedies', 'genre1_tv dramas', 'genre1_tv horror',
'genre1_tv sci-fi & fantasy', 'genre1_tv shows']

x=preprocessing.scale(dummies_df[columns])
y=preprocessing.scale(dummies_df['rating'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=seed)

import statsmodels.api as sm
from sklearn.metrics import mean_squared_error,mean_absolute_error

_x = sm.add_constant(pd.DataFrame(x_train, columns=columns))
est = sm.OLS(pd.DataFrame(y_train, columns=['price']), _x)
est = est.fit()
est.summary()
```

```python
_=pd.DataFrame(x_test, columns=columns)
_['const']=1
y_pred = est.predict(exog=_)

mse =mean_squared_error(y_test, y_pred,squared=True)
rmse =mean_squared_error(y_test, y_pred,squared=False)
mae=mean_absolute_error(y_test, y_pred)
print(f'rmse:{rmse}\nmae:{mae}\nmse:{mse}')
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

res_list=['years_to_add', 'oscar_number','duration_min', 'duration_season']

for i in res_list:
    fig = plt.figure(figsize=(12,8))
    fig = sm.graphics.plot_regress_exog(est, i, fig=fig)
```

```python
"""
Percebemos que apesar o MAE/MSE estarem muito bons, os resíduos não estão controlados.
Indicando que precisamos colocar mais features.
Perceba também que nem todas variáveis são significativas(p<0.05), portanto devem ser removidas .
Talvez o modelo se comporte melhor para uma faixa de duração de season >4 e years_to_add>2
"""
```

### Método não paramétrico

```python
"""
dado que farei um CV, um split não é obrigatório. 
"""
from sklearn.model_selection import train_test_split
dummies_df_=dummies_df[dummies_df['rating']!=0]

x=dummies_df_.iloc[:,2:]
y=dummies_df_.iloc[:,:1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=seed)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import make_scorer

n_estimators=[20,50,60,100,120,140]
max_features=['auto','sqrt','log2']
max_depth=[2,5,9,13,18,24,26]
min_samples_split=[2,5,7,10,12,15]
min_samples_leaf=[6,8,12,18]


scoring = {'mean_squared_error':make_scorer(mean_squared_error,greater_is_better=False)
           ,'mean_absolute_error':make_scorer(mean_absolute_error,greater_is_better=False)}

##if scoring: refit = False


paramns={'n_estimators':n_estimators,
'max_features':max_features,
'max_depth':max_depth,
'min_samples_split':min_samples_split,
'min_samples_leaf':min_samples_leaf
}
rfg = RandomForestRegressor(random_state=seed)
rfg_cv=RandomizedSearchCV(estimator=rfg
                         ,param_distributions=paramns
                         ,scoring=scoring
                         ,cv=10
                         ,refit=False
                         ,n_jobs=-1
                         ,random_state=seed)

rfg_cv = rfg_cv.fit(x_train,y_train)
```

```python
print('MSE',rfg_cv.cv_results_['rank_test_mean_squared_error'], rfg_cv.cv_results_['mean_test_mean_squared_error'][5])
print('MAE',rfg_cv.cv_results_['rank_test_mean_absolute_error'], rfg_cv.cv_results_['mean_test_mean_absolute_error'][5])
"""
no indice 5 para os dois casos 
"""
```

```python
rfg_cv.cv_results_['mean_test_mean_squared_error']
rfg_cv.cv_results_['mean_test_mean_absolute_error']

"""
não parece estar overfittando
"""
```

```python
rmse=mean_squared_error(y_test, rfg_cv.predict(x_test),squared=False)
mse=mean_squared_error(y_test, rfg_cv.predict(x_test),squared=True)

mae=mean_absolute_error(y_test, rfg_cv.predict(x_test))
print(f'mse :{mse}\nrmse:{rmse}\nmae :{mae}')

"""
O MSE está muito bom, com a vantagem de não ficarmos com a dúvida nos residuos. 
Agora é colocar em produção e ver se drifta.
"""
```
